```python
# Import the biden_tweets.csv file into a DataFrame.    
import pandas as pd
trump_df = pd.read_csv('trump_tweets.csv')
```

```python
trump_df.head()
```

```python
# drop the columns that are not needed
trump_df = trump_df.drop(columns=['author_id', 'edit_history_tweet_ids', 'id'])
```

```python
# Split the 'public_metrics' column into separate columns
def split_metrics(df):
    df = df.join(df['public_metrics'].apply(lambda x: pd.Series(eval(x))))
    df = df.drop(columns=['public_metrics'])

    return df
```

```python
trump_df = split_metrics(trump_df)
```

```python
trump_df
```

```python
import re
# Clean the tweet column

# Remove the URLs from the tweets
# Remove @mentions and #hashtags
# Remove special characters
# Remove RT
# Remove \n

def clean_tweet(tweet):
    pattern = r'@\w+|#\w+'
    emojis = r'[^\w\s,.!?]'
    tweet = re.sub(r'http\S+', '', tweet)
    tweet = re.sub(pattern, '', tweet)  
    tweet = re.sub(r'RT', '', tweet)
    tweet = re.sub(emojis, '', tweet)
    tweet = re.sub(r'\n+', ' ', tweet) 
    return tweet
```

```python
# Run the clean_tweet function on the 'text' column
trump_df['text'] = trump_df['text'].apply(clean_tweet)
```

```python
# Convert the 'created_at' column to a datetime object
def convert_created_at(df):
    df['created_at'] = pd.to_datetime(df['created_at'])
    return df
```

```python
# Run the convert_created_at function on the 'created_at' column
trump_df = convert_created_at(trump_df)
```

```python
#Output cleaned trump tweets to a new csv file
with open('clean_trump_tweets.csv', 'w') as f:
    trump_df.to_csv(f, index=False)
```
