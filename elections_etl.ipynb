```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Read data into a pandas dataframe
election_df = pd.read_csv('Resources/1976-2020-president.csv')

# Display dataframe
election_df.head()
```

```python
# Get information on the dataframe
election_df.info()
```

```python
# Drop unnecessary columns
election_df = election_df.drop(columns=['state_fips', 'state_cen', 'state_ic', 'office', 'writein', 'version', 'notes', 'party_simplified'])

# Display dataframe
election_df.head()
```

```python
# Drop rows with NaN values
election_df = election_df.dropna()

# Display dataframe
election_df.head()
```

```python
# Get information on the dataframe
election_df.info()
```

```python
# Create dataframe with data only from 2016
election_2016_df = election_df[(election_df['year'] == 2016)]

# Display dataframe
election_2016_df.head()
```

```python
# Create dataframe with data only from 2020
election_2020_df = election_df[(election_df['year'] == 2020)]

# Display dataframe
election_2020_df.head()
```

```python
# Create dataframe with data only from 2016 and 2020
election_2016_2020_df = election_df[(election_df['year'] == 2016) | (election_df['year'] == 2020)]

# Display dataframe
election_2016_2020_df.head()
```

```python
# Save dataframes to csv files
election_2016_df.to_csv('Resources/election_2016.csv', index=False)
election_2020_df.to_csv('Resources/election_2020.csv', index=False)
election_2016_2020_df.to_csv('Resources/election_2016_2020.csv', index=False)
```
